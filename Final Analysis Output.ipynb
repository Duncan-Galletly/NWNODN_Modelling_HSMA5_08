```python
import sys
import importlib
import subprocess
import gc

def install_and_check_libraries(lib_list):
    for lib in lib_list:
        try:
            # Try to import library
            importlib.import_module(lib)
            print(f"'{lib}' is already installed.")
        except ImportError:
            # If not installed, install library
            try:
                print(f">>>> Installing {lib}.")
                subprocess.check_call([sys.executable, "-m", "pip", "install", lib])
            except Exception as e:
                print(f"Error installing {lib}:", str(e))

libraries = ["pandas", "numpy", "openpyxl", "matplotlib", "scipy", "seaborn"]
install_and_check_libraries(libraries)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import matplotlib.patches as mpatches

# Load the data
file_path = './Solution_TT_Output.xlsx'
data = pd.read_excel(file_path)

regions_list = data['Regional Objectives'].unique().tolist()

for region_name in regions_list:

    print(region_name)
    # Filter data for a specific region, e.g., "Cheshire and Merseyside"
    regional_data = data[data['Regional Objectives'] == region_name]

    # Group by Solution_Number, Site, IMD_Decile, and summarize
    regional_summary = regional_data.groupby(['Solution_Number', 'Site', 'IMD_Decile']).agg(
        Mean_Adjusted_TT=('Adjusted_TT', 'mean')
    ).reset_index()

    # Prepare data for comparison by pivoting the table
    pivot_regional_summary = regional_summary.pivot_table(
        index=['Site', 'IMD_Decile'],
        columns='Solution_Number',
        values='Mean_Adjusted_TT'
    )

    # Flatten the columns after pivot to remove MultiIndex if necessary
    pivot_regional_summary.columns = pivot_regional_summary.columns.get_level_values(0)

    # Calculate average travel times for each solution
    solution_averages = pivot_regional_summary.mean()

    # Rank the solutions based on average travel times
    solution_ranks = solution_averages.rank(method='min').astype(int)

    # Calculate the difference of each solution from the others within the same site and IMD_Decile
    mean_adjusted_tts = pivot_regional_summary.mean(axis=1)
    solution_deviations = pivot_regional_summary.subtract(mean_adjusted_tts, axis=0)

    # Generate detailed insights for each solution
    detailed_summaries = {}
    for solution in solution_deviations.columns:
        solution_data = solution_deviations[solution].dropna()
        top_3_most_deviant = solution_data.nlargest(5)
        top_3_least_deviant = solution_data.nsmallest(5)

        # Get actual average travel times for the top and bottom 3
        actual_times_top = pivot_regional_summary.loc[top_3_most_deviant.index, solution]
        actual_times_bottom = pivot_regional_summary.loc[top_3_least_deviant.index, solution]
        
        # Statistical test: Compare each solution's mean travel time to the overall mean
        t_stat, p_value = stats.ttest_1samp(pivot_regional_summary[solution].dropna(), pivot_regional_summary.mean(axis=1).mean())

        # Determine significance
        significance = "not significant" if p_value >= 0.05 else "significant"

        # Trend analysis for IMD and site, assuming data is accessible
        by_imd = pivot_regional_summary.reset_index()
        trend_imd = by_imd.groupby('IMD_Decile')[solution].mean().corr(by_imd['IMD_Decile'])
        trend_type_imd = "shorter" if trend_imd > 0 else "longer"

        by_site = by_imd.groupby('Site')[solution].mean()
        site_trends = by_site.sort_values(ascending=False).index.tolist()

        # Generate summary including trend analysis
        detailed_summaries[solution] = {
            'average_time': solution_averages[solution],
            'rank': solution_ranks[solution],
            'trend_imd': trend_type_imd,
            'correlation_imd': trend_imd,
            'top_sites': site_trends,
            'significance': significance,
            't_stat': t_stat,
            'p_value': p_value
        }

        # Output the results
        details = detailed_summaries[solution]
            
        print(f"Solution {solution} Insights:")
        print(f"Average travel time: {details['average_time']:.2f} minutes, ranked {details['rank']} among all solutions.")
        print(f"Trend in IMD Deciles: Travel times are {details['trend_imd']} at lower IMD deciles, correlation coefficient {details['correlation_imd']:.2f}.")
        print("Top 5 most negatively affected IMD Deciles with Actual Times:")
        for idx, value in actual_times_top.items():
            print(f" - {idx[0]} in IMD decile {idx[1]} with a mean travel time of {value:.2f} minutes.")
        print("Top 5 most positively affected IMD Deciles with Actual Times:")
        for idx, value in actual_times_bottom.items():
            print(f" - {idx[0]} in IMD decile {idx[1]} with a mean travel time of {value:.2f} minutes.")
        print(f"The difference from the overall mean is statistically {details['significance']} (t-statistic = {details['t_stat']:.2f}, p-value = {details['p_value']:.3f}).")
        print("\n")
        
        # Define the figure and plot the data
        plt.figure(figsize=(7, 2))
        plt.rcParams["font.family"] = "Arial"
        plt.title(f'Mean Adjusted Travel Times for Solution {solution}', horizontalalignment = 'right', pad=15)
        concatenated_times = pd.concat([actual_times_top, actual_times_bottom])
        colors = ['gold' if idx in top_3_most_deviant else 'teal' for idx in concatenated_times.index]
        ax = concatenated_times.plot(kind='bar', color=colors)
        plt.xlabel('(Site, IMD Decile)', labelpad = 10, loc='right')
        plt.xticks(rotation=45, ha='right')
        plt.ylabel('Mean Adjusted Travel Time')
        solution_average = pivot_regional_summary[solution].mean()
        plt.axhline(solution_average, color='grey', linewidth=0.8)
        for idx, val in enumerate(concatenated_times):
            plt.text(idx, val + 0.05, f'{val:.2f}', ha='center', va='bottom', color='black', size=8)
        for spine in ax.spines.values():
            spine.set_visible(False)
        gold_patch = mpatches.Patch(color='gold', label='Most Negatively Affected')
        teal_patch = mpatches.Patch(color='teal', label='Most Positively Affected')
        plt.legend(handles=[gold_patch, teal_patch], loc='upper right' , bbox_to_anchor=(1,1.25))

        plt.show()
```

```python
# Load the data
file_path = './Solution_Activity_Output.xlsx'
activity_data = pd.read_excel(file_path)


regions_list = data['Regional Objectives'].unique().tolist()

for region_name in regions_list:

    print(region_name)

    # Data preparation and filtering for a specific region
    filtered_activity_data = activity_data[activity_data['Regional Objectives'] == region_name]

    # Group data by Solution_Number and Site, then calculate mean values
    grouped_data = filtered_activity_data.groupby(['Solution_Number', 'Site']).agg({
        'Adjusted_Count': 'mean',
        'AVG_NICU_Bed_Numbers': 'mean'
    }).reset_index()

    # Calculate overall averages for each site to compare against individual solution performance
    site_averages = grouped_data.groupby('Site').agg({
        'Adjusted_Count': 'mean',
        'AVG_NICU_Bed_Numbers': 'mean'
    })

    # Initialize lists to collect all deviations
    all_count_deviations = []
    all_bed_deviations = []

    # Calculate deviations and collect them
    for _, row in grouped_data.iterrows():
        site_avg_count = site_averages.loc[row['Site'], 'Adjusted_Count']
        site_avg_beds = site_averages.loc[row['Site'], 'AVG_NICU_Bed_Numbers']

        count_deviation = (row['Adjusted_Count'] - site_avg_count) / site_avg_count * 100
        beds_deviation = (row['AVG_NICU_Bed_Numbers'] - site_avg_beds) / site_avg_beds * 100

        all_count_deviations.append(count_deviation)
        all_bed_deviations.append(beds_deviation)

    # Calculate the quartile threshold for significant deviations
    count_deviation_threshold = np.percentile(all_count_deviations, 75)
    bed_deviation_threshold = np.percentile(all_bed_deviations, 75)

    # Detailed insights for each solution with significance based on top quartile
    for solution in grouped_data['Solution_Number'].unique():
        solution_data = grouped_data[grouped_data['Solution_Number'] == solution]
        
        significant_findings = False  # Flag to check if any significant findings exist
        significant_details = []  # To store significant findings

        for _, row in solution_data.iterrows():
            site_avg_count = site_averages.loc[row['Site'], 'Adjusted_Count']
            site_avg_beds = site_averages.loc[row['Site'], 'AVG_NICU_Bed_Numbers']

            count_deviation = (row['Adjusted_Count'] - site_avg_count) / site_avg_count * 100
            beds_deviation = (row['AVG_NICU_Bed_Numbers'] - site_avg_beds) / site_avg_beds * 100

            significant_count_deviation = abs(count_deviation) > count_deviation_threshold
            significant_beds_deviation = abs(beds_deviation) > bed_deviation_threshold

            if significant_count_deviation or significant_beds_deviation:
                significant_findings = True
                details = f"  Site: {row['Site']}\n"
                details += f"    Adjusted Activity: {row['Adjusted_Count']:.2f}, Deviation: {count_deviation:.2f}%\n"
                details += f"    NICU Beds: {row['AVG_NICU_Bed_Numbers']:.2f}, Deviation: {beds_deviation:.2f}%\n"
                significant_details.append(details)
        
        if significant_findings:
            print(f"Solution {solution} Significant Deviations:")
            for detail in significant_details:
                print(detail)
            print("\n")
```

```python
# Load the data from the uploaded Excel file
file_path = './Solution_1_Aggregated.xlsx'
sol1_data = pd.read_excel(file_path)
```

```python
# regions = ['Cheshire and Merseyside','Healthier Lancashire and South Cumbria','Greater Manchester Health and Social Care Partnership']

# sol1_data = sol1_data[sol1_data['STP_Name'].isin(regions)]

sol1_data_NICU = sol1_data[sol1_data['CC_Level'] == 'NICU']
```

```python
original_data = sol1_data_NICU[sol1_data_NICU['Solution_Number'] == 0]
modelled_data = sol1_data_NICU[sol1_data_NICU['Solution_Number'] != 0]

original_data_summed = original_data.groupby(['Site', 'Year']).agg({
    'Avg_Travel_Time': 'mean',
    'Sum_Record_Count': 'sum'
}).reset_index()

modelled_data_summed = modelled_data.groupby(['Site', 'Year', 'Algorithm',	'Weighted',	'OI_Inc']).agg({
    'Avg_Travel_Time': 'mean',
    'Sum_Record_Count': 'sum'
}).reset_index()

original_data_grouped = original_data_summed.groupby(['Site']).agg({
    'Avg_Travel_Time': 'mean',
    'Sum_Record_Count': 'mean'
}).reset_index()

modelled_data_grouped = modelled_data_summed.groupby(['Site']).agg({
    'Avg_Travel_Time': 'mean',
    'Sum_Record_Count': 'mean'
}).reset_index()
```

```python
# Merge the original data with modelled data for comparison
comparison_data = pd.merge(original_data_grouped, modelled_data_grouped, on=['Site'], suffixes=('_Orig', '_Mod'))

# Fitting polynomial curves (let's use a 2nd degree polynomial for example)
poly_fit_time = np.polyfit(comparison_data['Avg_Travel_Time_Orig'], comparison_data['Avg_Travel_Time_Mod'], 2)
poly_fit_count = np.polyfit(comparison_data['Sum_Record_Count_Orig'], comparison_data['Sum_Record_Count_Mod'], 2)

# Generating polynomial curve data
poly_line_time = np.poly1d(poly_fit_time)
poly_line_count = np.poly1d(poly_fit_count)

# Points for curve
x_time = np.linspace(comparison_data['Avg_Travel_Time_Orig'].min(), comparison_data['Avg_Travel_Time_Orig'].max(), 100)
x_count = np.linspace(comparison_data['Sum_Record_Count_Orig'].min(), comparison_data['Sum_Record_Count_Orig'].max(), 100)

# Plotting
fig, (ax2, ax1) = plt.subplots(1, 2, figsize=(12, 5))
sns.scatterplot(data=comparison_data, x='Avg_Travel_Time_Orig', y='Avg_Travel_Time_Mod', hue='Site', ax=ax1)
sns.scatterplot(data=comparison_data, x='Sum_Record_Count_Orig', y='Sum_Record_Count_Mod', hue='Site', ax=ax2)
ax1.plot(x_time, poly_line_time(x_time), color="red", label='Fit')
ax2.plot(x_count, poly_line_count(x_count), color="red", label='Fit')
ax1.set_ylim(ax1.get_xlim())
ax2.set_ylim(ax2.get_xlim())
ax1.set_title('Comparison of Average Travel Times by Site')
ax1.set_xlabel('Original Avg Travel Time')
ax1.set_ylabel('Modelled Avg Travel Time')
ax2.set_title('Comparison of NICU Activity By Site')
ax2.set_xlabel('Original Sum Record Count')
ax2.set_ylabel('Modelled Sum Record Count')

plt.legend([],[], frameon=False)
ax1.legend(loc='upper center', bbox_to_anchor=(1.5, 1),
           fancybox=True, shadow=True, ncol=1)
ax2.legend().set_visible(False)
plt.tight_layout()
plt.show()
```

```python

original_site_sizes = original_data_grouped.groupby('Site').agg({'Sum_Record_Count': 'sum'}).reset_index()
modelled_site_sizes = modelled_data_grouped.groupby('Site').agg({'Sum_Record_Count': 'sum'}).reset_index()

# Calculating standard deviation
std_dev_original = original_site_sizes['Sum_Record_Count'].std()
std_dev_modelled = modelled_site_sizes['Sum_Record_Count'].std()

# Visualizing site size distributions
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
sns.histplot(original_site_sizes['Sum_Record_Count'], kde=True, ax=ax1, color='blue', label='Original')
sns.histplot(modelled_site_sizes['Sum_Record_Count'], kde=True, ax=ax2, color='red', label='Modelled')

ax1.set_title(f'Original Site Sizes Distribution\nStd Dev: {std_dev_original:.2f}')
ax1.set_xlabel('Sum of Record Count')
ax1.set_ylabel('Frequency')
ax1.legend()

ax2.set_title(f'Modelled Site Sizes Distribution\nStd Dev: {std_dev_modelled:.2f}')
ax2.set_xlabel('Sum of Record Count')
ax2.set_ylabel('Frequency')
ax2.legend()

plt.tight_layout()
plt.show()
```
